```python
# Programming Assignment Q2

import numpy as np
import matplotlib.pyplot as plt
```

```python
# Use np.linspace to create a grid of 50 points ranging from 0 to 1
x = np.linspace(0,1,50)

# Create a numpy array  ùë¶  containing the values for the natural logarithm function using the np.log(x) function.
# Create a numpy array  ùëß  containing the values for the exponential function using the np.exp(x) function.
y = np.log(x)
z = np.exp(x)
```

```python
# Use the plt.scatter(x,y) method for the  ùë¶  and  ùëß  vectors to create two scatter plots of the points you've created.
# Use the plt.show() method to create the plot. That plot has some problems.
plt.scatter(x,y)
plt.scatter(x,z)
plt.show()
```

```python
# Before the plt.show() call, add labels to the  ùë•  and  ùë¶  axes using plt.xlabel(label) and plt.ylabel(label).
  # Add a title to the graph using plt.title(title), like "Natural Log and Exponential Functions".
plt.scatter(x,y)
plt.scatter(x,z)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Natural Log and Exponential Functions')
plt.show()
```

```python
# That looks a lot better, but we need a legend. When you screate the scatter plots,
  # add the argument label='Natural Log' or label='Exponential' to the .scatter method call.
  # Before the plt.show() call, add a plt.legend(loc = 'lower right') method call, which creates a legend in the lower right.

plt.scatter(x,y, label='Natural Log')
plt.scatter(x,z, label='Exponential Log')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Natural Log and Exponential Functions')
plt.legend(loc = 'lower right')
plt.show()
```

```python
# Create a grid of 100 equally spaced points ranging from -6.5 to 6.5.
x = np.linspace(-6.5,6.5,100)

# Use the sine and cosine functions in Numpy to compute the values of those functions for each point on your grid. (You'll have to find out what those functions are.)
y = np.sin(x)
z = np.cos(x)

# Plot the values of the two functions for the values on the grid on the same plot.
plt.scatter(x,y)
plt.scatter(x,z)
plt.show()

# The scatter plot looks really noisy. Instead of plt.scatter(x,y) to make a scatter plot, use plt.plot(x,y) to make a line graph.
plt.plot(x,y)
plt.plot(x,z)
plt.show()

# Make the plot again, with labels for the axes, a title, and a legend in the lower left instead of the lower right.
plt.plot(x,y,label='Sine')
plt.plot(x,z,label='Cosine')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Sine and Cosine')
plt.legend(loc = 'lower right')
plt.show()
```
